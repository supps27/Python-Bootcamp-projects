```python
import random
from high_low_list_dict_dataset_1000 import high_low_list_dict_dataset

def get_random_question():
    """Returns a random (question, answer) pair from the dataset, with integer answer."""
    item = random.choice(high_low_list_dict_dataset)
    return item["question"], int(item["answer"])  # Convert answer to integer

def get_user_guess():
    """Gets user's guess as an integer."""
    guess = int(input("Enter your guess (integer): "))
    return guess

def check_guess(guess, answer):
    """
    Compares guess with the correct answer.
    Returns 'correct', 'low', or 'high'.
    """
    if guess == answer:
        return "correct"
    elif guess < answer:
        return "low"
    else:
        return "high"

def play_one_round():
    """Plays a single round of the high-low game."""
    question, answer = get_random_question()
    print(f"\n🤔 Question: What is the {question}?")

    guess = get_user_guess()
    result = check_guess(guess, answer)

    if result == "correct":
        print("🎯 Exactly right!")
    elif result == "low":
        print("📉 Too low!")
    elif result == "high":
        print("📈 Too high!")

    return result != "correct"  # True if guess was wrong

def play_game():
    """Main game loop that tracks score and continues rounds."""
    score = 0
    print("🎮 Welcome to the High-Low Guessing Game!")

    while True:
        try:
            wrong = play_one_round()
        except ValueError:
            print("❌ Invalid input! Please enter an integer.")
            continue

        score += 1
        if wrong:
            print("❌ Wrong guess! Game over.")
            break

        cont = input("Do you want to keep playing? (y/n): ").strip().lower()
        if cont != "y":
            break

    print(f"\n🏁 Final Score: {score}")

# 🔁 Start the game
play_game()
```

```python
import random
from high_low_comparison_dataset_1000 import high_low_comparison_dataset as dataset  # use your valid file/module name

def get_two_random_choices():
    """Get two unique random data points from the dataset."""
    datapoint1 = random.choice(dataset)
    datapoint2 = random.choice(dataset)

    while datapoint1 == datapoint2:  # ensure they are different
        datapoint2 = random.choice(dataset)

    return datapoint1, datapoint2

def get_user_input(name1, name2, category):
    """Ask user to guess whether 2nd is Higher or Lower than 1st."""
    print(f"\nA: {name1} has higher {category} or lower {category} than {name2}?")
    # print(f"A: {name1}")
    # print(f"B: {name2}")
    response = input("Enter 'Higher' or 'Lower': ").strip().lower()
    return response

def compare(datapoint1, datapoint2, user_guess):
    """Compare the values and check user guess."""
    val1 = datapoint1["value"]
    val2 = datapoint2["value"]

    if (val2 > val1 and user_guess == "higher") or (val2 < val1 and user_guess == "lower"):
        print("✅ Correct! You guessed it right.")
        return True
    else:
        print("❌ Wrong! Game Over.")
        print(f"{datapoint2['name']} has {datapoint2['value']} vs {datapoint1['value']}")
        return False

# 🎮 Main game loop
def play_high_low_game():
    score = 0
    while True:
        dp1, dp2 = get_two_random_choices()
        category = dp1["category"]
        guess = get_user_input(dp1["name"], dp2["name"], category)

        if compare(dp1, dp2, guess):
            score += 1
            print(f"Your Score: {score}")
        else:
            break

play_high_low_game()
```
